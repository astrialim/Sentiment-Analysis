```python
import re
def casefolding(Review_Text):
  Review_Text = Review_Text.lower()
  Review_Text = Review_Text.strip()
  Review_Text = re.sub(r'[,.?!;:/-123(")²]', '', Review_Text)
  return Review_Text
df['Review_Text'] = df['Review_Text'].apply(casefolding)
df
```

```python
def token(Review_Text):
  dele = Review_Text.split(' ')
  dat = []
  a = -1
  for hu in dele:
    a += 1
  if hu == '':
    dat.append(a)
  p = 0
  b = 0
  for q in dat:
    b = q - p
    del dele[b]
    p += 1
  return dele
df['Review_Text'] = df['Review_Text'].apply(token)
df
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
def stopword_removal(Review_Text):
  filtering = stopwords.words('indonesian', 'english')
  x = []
  d = []
  def func(x):
    if x in filtering:
      return False
    else:
      return True
  fil = filter(func, Review_Text)
  for x in fil:
    d.append(x)
  return d
df['Review_Text'] = df['Review_Text'].apply(stopword_removal)
df
```

```python
def stemming(Review_Text):
  facto = stmf()
  stemmer = facto.create_stemmer()
  e = []
  for i in Review_Text:
    f = stemmer.stem(i)
    e.append(f)
  d_clean = []
  d_clean = ' '.join(e)
  print(d_clean)
  return d_clean
df['Review_Text'] = df['Review_Text'].apply(stemming)
```

```python
df.to_csv('clean.csv', index=False)
clean = pd.read_csv('clean.csv')
clean
```

```python
clean = clean.astype({'Sentiment' : 'category'})
clean = clean.astype({'Star_Rating' : 'category'})
clean = clean.astype({'Review_Text' : 'string'})
clean.dtypes
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer()
tf_text = tf.fit_transform(clean['Sentiment'].astype('U'))
tf_text
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(tf_text, df['Sentiment'], test_size=0.2, random_state=22)

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
exam = MultinomialNB().fit(x_train, y_train)
pred = exam.predict(x_test)
print('Accuracy : ', accuracy_score(y_test, pred))
print('Precision : ', precision_score(y_test, pred, average='weighted'))
print('Recall : ', recall_score(y_test, pred, average='weighted'))
print('F1 Score : ', f1_score(y_test, pred, average='weighted'))

print(f'Confusion Matrix :\n {confusion_matrix(y_test, pred)}')
print('-------------------------------------------------------\n')
print(classification_report(y_test, pred, zero_division=0))
```
